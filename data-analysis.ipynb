```python
import pandas as pd
```

```python
df = pd.read_csv('/content/canadian_immegration_data 2.csv')
```

```python
df
```

```python
df.info()
```

```python
df.set_index('Country', inplace=True)
```

```python
year = list(map(str, range(1980, 2014)))
```

```python
year
```

```python
brazil = df.loc['Brazil', year]
```

```python
brazil
```

```python
brazil_dict={'year': brazil.index.tolist(), 'immigrants': brazil.values.tolist()}
data_brazil = pd.DataFrame(brazil_dict)
```

```python
data_brazil
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(data_brazil['year'], data_brazil['immigrants'], lw=3, marker='o', color='tab:red')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('\nImmigration from Brazil to Canada \nfrom 1980 to 2013\n\n', fontsize=18, loc='left')
ax.set_xlabel('\nYear\n', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.set_ylabel('\nNumber of Immigrants\n', fontsize=14)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.savefig('brazil_to_canada_immigration.png', transparent=True, dpi=300, bbox_inches='tight')
plt.show()
```

```python
fig,axs = plt.subplots(1, 2, figsize=(15,5))
axs[0].plot(data_brazil['year'], data_brazil['immigrants'], lw=3, color='tab:red')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_title('Immigration from Brazil to Canada \n1980 to 2013')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Number of Immigrants')
axs[0].grid()

axs[1].boxplot(data_brazil['immigrants'])
axs[1].set_title('Boxplot of Immigrants from Brazil to Canada \n1980 to 2013')
axs[1].set_xlabel('Brazil')
axs[1].set_ylabel('Number of Immigrants')
axs[1].grid()
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Immigration from the four largest countries in South America')

axs[0,0].plot(df.loc['Brazil', year])
axs[0,0].set_title('Immigration from Brazil to Canada')

axs[0,1].plot(df.loc['Colombia', year])
axs[0,1].set_title('Immigration from Colombia to Canada')

axs[1,0].plot(df.loc['Argentina', year])
axs[1,0].set_title('Immigration from Argentina to Canada')

axs[1,1].plot(df.loc['Peru', year])
axs[1,1].set_title('Immigration from Peru to Canada')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
  ax.set_xlabel('Year')
  ax.set_ylabel('Number of Immigrants')
  ax.grid()

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim([ymin, ymax])

plt.show()
```

```python
south_america = df.query('Region == "South America"')
```

```python
south_america
```

```python
south_america_sorted = south_america.sort_values('Total', ascending=True)
```

```python
colors = []
for country in south_america_sorted.index:
  if country == 'Brazil':
    colors.append('tab:green')
  else:
    colors.append('silver')


fig, ax = plt.subplots(figsize=(12,5))
ax.barh(south_america_sorted.index, south_america_sorted['Total'], color=colors)
ax.set_title('\n\nSouth America: Brazil is the fourth largest country with the most immigrants to Canada \nfrom 1980 to 2013\n', fontsize=18)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(south_america_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left',va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('south_america_immigration.png', transparent=False, dpi=300, bbox_inches='tight')
plt.show()

plt.show()
```

```python
fig,ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set_title('\nCountries with the highest immigration to Canada \n1980 to 2013\n', loc='left', fontsize=16)
ax.set_xlabel('\nNumber of Immigrants\n', fontsize=14)
ax.set_ylabel('')

plt.show()
```

```python
def chose_palette(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette)
  ax.set_title('\nCountries with the highest immigration to Canada \n')
  ax.set_xlabel('\nNumber of Immigrants\n', fontsize=14)
  ax.set_ylabel('')
  plt.show()
```
